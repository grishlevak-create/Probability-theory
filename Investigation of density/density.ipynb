```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
a, b = 1, 5
dist = sts.uniform(loc=a, scale=b-a)

sample = dist.rvs(size=10000)

print("РАВНОМЕРНОЕ РАСПРЕДЕЛЕНИЕ")
print(f"Теория: E[X] = {(a+b)/2}, D[X] = {(b-a)**2/12}")
print(f"Выборка: E[X] = {np.mean(sample):.4f}, D[X] = {np.var(sample):.4f}")

x = np.linspace(a-1, b+1, 300)
plt.plot(x, dist.pdf(x))
plt.title("Плотность равномерного распределения")
plt.grid()
plt.show()
```

```python
lam = 2
dist = sts.expon(scale=1/lam)

sample = dist.rvs(size=10000)

print("ЭКСПОНЕНЦИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ")
print(f"Теория: E[X] = {1/lam}, D[X] = {1/lam**2}")
print(f"Выборка: E[X] = {np.mean(sample):.4f}, D[X] = {np.var(sample):.4f}")

x = np.linspace(0, 5, 300)
plt.plot(x, dist.pdf(x))
plt.title("Плотность экспоненциального распределения")
plt.grid()
plt.show()
```

```python
mu, sigma = 1, 2
dist = sts.norm(loc=mu, scale=sigma)

sample = dist.rvs(size=10000)

print("НОРМАЛЬНОЕ РАСПРЕДЕЛЕНИЕ")
print(f"Теория: E[X] = {mu}, D[X] = {sigma**2}")
print(f"Выборка: E[X] = {np.mean(sample):.4f}, D[X] = {np.var(sample):.4f}")

x = np.linspace(mu-4*sigma, mu+4*sigma, 300)
plt.plot(x, dist.pdf(x))
plt.title("Плотность нормального распределения")
plt.grid()
plt.show()
```

```python
nu = 5
dist = sts.t(df=nu)

sample = dist.rvs(size=10000)

print("РАСПРЕДЕЛЕНИЕ СТЬЮДЕНТА")
print("Теория: E[X] = 0")
print(f"Теория: D[X] = {nu/(nu-2):.4f}")
print(f"Выборка: E[X] = {np.mean(sample):.4f}")
print(f"Выборка: D[X] = {np.var(sample):.4f}")

x = np.linspace(-6, 6, 400)
plt.plot(x, dist.pdf(x))
plt.title("Плотность распределения Стьюдента")
plt.grid()
plt.show()
```
