p(A) = ((10!/3!7!) + (3!/3!0!) + (7!/3!4!)) / (20!/3!17!) = 0,1368

В одном испытании событие происходит с вероятностью p.  Не происходит с 1 - p.

Вероятность хотя бы одного успеха при n испытаниях:  1 - (1 - p)^n ≥ P

(1 - p)^n ≤ 1 - P

n ≥ ln(1 - P) / ln(1 - p)

Подставим p = 0.4, P = 0.8704:

1 - p = 0.6 

1 - P = 0.1296

ln(0.1296) ≈ -2.043

ln(0.6) ≈ -0.511

n ≥ (-2.043) / (-0.511) ≈ 3.997

n = 4
