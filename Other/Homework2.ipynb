```python
import random

def simulate():
    #  1: 0 - , 1 - 
    urn1 = [0]*15 + [1]*5
    #  2: 0 - , 1 - 
    urn2 = [0]*6 + [1]*1
    
    #  3   1  2
    chosen = random.sample(urn1, 3)
    for ball in chosen:
        urn2.append(ball)
    
    #   1   2
    drawn = random.choice(urn2)
    return drawn == 0  # True  

#   
N = 1000000
white_count = 0
for _ in range(N):
    if simulate():
        white_count += 1

f" : {white_count / N:.6f}"
```
