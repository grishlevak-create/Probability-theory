# Задача
Решить классически и смоделировать решение задачи. В одной урне 15 белых и 5
черных шаров, а в другой - 6 белых и 1 черный шар. Из первой урны случайным
образом вынимают 3 шара и опускают во вторую урну. После этого из второй урны
также случайно вынимают 1 шар. Найти вероятность того, что шар, вынутый из второй
урны - белый. 

```python
import random

def simulate():
    # Урна 1: 0 - белый, 1 - чёрный
    urn1 = [0]*15 + [1]*5
    # Урна 2: 0 - белый, 1 - чёрный
    urn2 = [0]*6 + [1]*1
    
    # Перекладываем 3 шара из урны1 в урну2
    chosen = random.sample(urn1, 3)
    for ball in chosen:
        urn2.append(ball)
    
    # Случайно вынимаем 1 шар из урны2
    drawn = random.choice(urn2)
    return drawn == 0  # True если белый

# Моделируем много раз
N = 1000000
white_count = 0
for _ in range(N):
    if simulate():
        white_count += 1

f"Смоделированная вероятность: {white_count / N:.6f}"
```
