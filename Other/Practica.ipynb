```python
import numpy as np
from scipy.stats import binom, poisson, norm
import matplotlib.pyplot as plt
from math import sqrt

np.set_printoptions(formatter={"float_kind":"{:0.4f}".format})
```

```python
# Параметры биномиального распределения
n = 10   # количество испытаний
p = 0.5  # вероятность успеха

# Создаем объект биномиального распределения
binomial_dist = binom(n, p)

# Создаем массив всех возможных значений m (от 0 до n)
m_values = np.arange(0, n + 1)

# Рассчитываем вероятности для каждого m
probabilities = binomial_dist.pmf(m_values)

# Выводим значения
print("Значения m:", m_values)
print("Вероятности P(m):", probabilities)

# Строим график распределения
plt.bar(m_values, probabilities, color='skyblue', edgecolor='black')
plt.title(f'Биномиальное распределение (n={n}, p={p})')
plt.xlabel('m — количество успехов')
plt.ylabel('P(m)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# Программная реализация 3.50

```python
import numpy as np
import pandas as pd

def mean(R):
    return sum(r*p for r, p in zip(R.keys(), R.values()))

X = {0: 0.1, 1: 0.6, 2: 0.3}
Y = {0: 0.5, 2: 0.5}

Z = dict()
for x, px in X.items():
    for y, py in Y.items():
        z = x + y
        Z[z] = Z.get(z, 0) + px * py

print(Z) 
print(f"M(X)={mean(X)}")
print(f"M(Y)={mean(Y)}")
print(f"M(Z)={mean(Z):.2f}, M(X)+M(Y) = {mean(X)+mean(Y)}")

Z = dict(sorted(Z.items()))

df = pd.DataFrame([Z], index=["p"])
df
```

# Прога для 3.53 Саня разбирал

```python
import numpy as np
import pandas as pd
from scipy.stats import binom 
def var(R):
    return sum(r**2*p for r, p in zip(R.keys(), R.values())) - mean(R)**2
def mean(R):
    return sum(r*p for r, p in zip(R.keys(), R.values())) 

n, p = 2, 0.6
X_d = binom(n, p)
X = {x: X_d.pmf(x) for x in range(n+1)}

n, p = 2, 0.7
Y_d = binom(n, p)
Y = {y: Y_d.pmf(y) for y in range(n+1)}

Z = dict()
for x, px in X.items():
    for y, py in Y.items():
        z = x + y
        Z[z] = Z.get(z, 0) + px * py

print(Z) 
print(f"M(X)={mean(X)}")
print(f"M(Y)={mean(Y)}")
print(f"M(Z)={mean(Z):.2f}, M(X)+M(Y) = {mean(X)+mean(Y)}")

Z = dict(sorted(Z.items()))

df = pd.DataFrame(sorted(Z.items()), columns=["z", "P(Z=z)"])
df
```
